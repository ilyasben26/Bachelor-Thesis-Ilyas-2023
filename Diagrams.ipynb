```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('runs.csv')
df
```

```python
# Create a figure with two subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# Create a scatter plot in the first subplot
df.plot(ax=axes[0], kind='scatter', x='vertices_count', y='render_time')
axes[0].set_xlabel('number of vertices')
axes[0].set_ylabel('render time in seconds')

# Create a scatter plot in the second subplot
df.plot(ax=axes[1], kind='scatter', x='faces_count', y='render_time')

axes[1].set_xlabel('number of faces')
axes[1].set_ylabel('render time in seconds')

# Show the plot
plt.savefig('two_plots.png')
plt.show()
```

```python
# Create a scatter plot from the dataframe
ax = df.plot(kind='scatter', x='batch_size', y='CLIP_time')

# Set the x-axis tick values and labels
ax.set_xticks([3, 6])
ax.set_xlim(0,9)
ax.set_xticklabels(['3', '6'])

# Set the x and y axis labels
ax.set_xlabel('Batch size')
ax.set_ylabel('CLIP time in seconds')

# Show the plot
plt.savefig('CLIP_time_batch.png')
plt.show()
```
